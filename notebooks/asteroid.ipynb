# Asteroid Close-Approach Data Exploration
This notebook explores near-Earth asteroid data as part of the Asteroid Close-Approach Explorer project.

**Steps:**
1. Load asteroid data
2. Explore data: distributions, top-10 closest, scatter plots
3. (Optional) Train a simple classifier

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load asteroid data
asteroids_df = pd.read_csv('../data/asteroids.csv')
asteroids_df.head()
```

```python
# Explore DataFrame structure and summary statistics
asteroids_df.info()
asteroids_df.describe()
```

```python
# Plot histogram for asteroid size (size_km)
plt.figure(figsize=(8,5))
sns.histplot(asteroids_df['size_km'], bins=30, kde=True)
plt.title('Distribution of Asteroid Sizes (km)')
plt.xlabel('Size (km)')
plt.ylabel('Count')
plt.show()

# Plot histogram for relative velocity (rel_vel_kps)
plt.figure(figsize=(8,5))
sns.histplot(asteroids_df['rel_vel_kps'], bins=30, kde=True, color='orange')
plt.title('Distribution of Relative Velocity (km/s)')
plt.xlabel('Relative Velocity (km/s)')
plt.ylabel('Count')
plt.show()
```
